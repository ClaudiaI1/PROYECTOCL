```python
import pandas as pd # Importamos la libreria pandas para manipulación de datos.
import matplotlib.pyplot as plt # Importamos la libreria matplotlib para visualizaicón de datos

# Cargamos los datos limpios.

data_path = "../data/processed/cleaned_products.csv" # Definimos la ruta del archivo de los datos a leer. 
df = pd.read_csv(data_path) # Cafgamos los datos del archivo csv en un DataFrame.

# Analisis y visualizacion

print("Basic Statistics")
print(df.describe()) # Mostramos un resumen estadistico de los datos en el DataFrame.

#nHistograma de precios.

plt.hist(df["price"], bins=20, edgecolor="black") # Creamos un histograma de la columna "Price" con 20 barras y bordes negros
plt.title("Distribucion de precios") # Añadimos una etiqueta al eje x
plt.ylabel("Precio") # Añadimos una etiqueta al eje x
plt.ylabel("Frecuenca") # Añadimos una etiqueta al eje Y
plt.show() # Mostramso el Histograma.

# Productos mas caros.

print("Productos mas Caros: ")
highestPrice= df.nlargest(5, 'price')
print(highestPrice)
```
